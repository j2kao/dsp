```python
import thinkstats2
import thinkplot
import nsfg
import math
import random
```

```python
def generate_rand_list(length):
    rand_list = [random.random() for _ in range(length)]
    return rand_list
```

```python
rands = generate_rand_list(1000)
```

```python
pmf = thinkstats2.Pmf(rands)
thinkplot.Pmf(pmf)
thinkplot.Show(xlabel='value', ylabel='PMF')
```

```python
cdf = thinkstats2.Cdf(rands)
thinkplot.Cdf(cdf)
thinkplot.Show(xlabel='value', ylabel='CDF')
```

* it's not uniform (smoother cdf and pmf), but it should approach uniform as length -> infinity
* can increase length to check

```python
new_rands = generate_rand_list(10000)
```

```python
pmf = thinkstats2.Pmf(new_rands)
thinkplot.Pmf(pmf)
thinkplot.Show(xlabel='value', ylabel='PMF')
```

```python
cdf = thinkstats2.Cdf(new_rands)
thinkplot.Cdf(cdf)
thinkplot.Show(xlabel='value', ylabel='CDF')
```

* getting better
